```python
def decode(x):
    global idlc
    prclid = str(x)[-6:]
    lc_type = str(x).split('0')[0]
    if prclid not in idlc:
        idlc[prclid] = [lc_type]
    else:
        idlc[prclid].append(lc_type)
```

```python
idlc={}
f = lambda x: decode(x)
[f(xi) for xi in ar]
```

```python
nums = [i for i in range(1,17)]
nums.remove(10)
```

```python
df = pd.DataFrame({'JOIN_INDEX':[],'Open Water':[], 'Developed, Open Space':[], 'Developed, Low Intensity':[],
'Developed, Medium Intensity':[], 'Developed, High Intensity':[], 
'Barren Land':[], 'Deciduous Forest':[], 'Evergreen Forest':[],
'Mixed Forest':[], 'Shrub/Scrub':[],  'Herbaceuous':[], 'Hay/Pasture':[],
'Cultivated Crops':[], 'Woody Wetlands':[], 'Emergent Herbaceuous Wetlands':[]})
```

```python
def pop_df(key):
    rec=[int(key)]
    for n in nums:
        rec.append(idlc[key].count(str(n))/len(idlc[key]))
    df.loc[len(df)] = rec
```

```python
f = lambda x: pop_df(x)
[f(ix) for ix in idlc]
```
